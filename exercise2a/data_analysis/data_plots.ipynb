```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

# Load the data
weak_scaling = pd.read_csv('../collected_data/weak_scaling_4thin.csv')

# Remove the size column
weak_scaling = weak_scaling.drop(columns='size')

# Plot the data
plt.figure(figsize=(12,  6))
sns.lineplot(x='processes', y='flat', data=weak_scaling, label='Flat',     markers=True, dashes=False)
sns.lineplot(x='processes', y='chain', data=weak_scaling, label='Chain',    markers=True, dashes=False)
sns.lineplot(x='processes', y='bintree', data=weak_scaling, label='BinTree', markers=True, dashes=False)


# Customize the plot
plt.title('Weak scaling')
plt.xlabel('Number of Processes')
plt.ylabel('Time (microseconds)')
plt.legend()

# Display the plot
plt.show()
```

```python
#read strong scaling data
strong_scaling = pd.read_csv('../collected_data/results_strong_scaling_2thin.csv')

# Plot the data
plt.figure(figsize=(12,  6))


sns.lineplot(x='size', y='flat', data=strong_scaling, label='flat',     markers=True, dashes=False, marker='o')
sns.lineplot(x='size', y='chain', data=strong_scaling, label='chain',    markers=True, dashes=False, marker='o')
sns.lineplot(x='size', y='bintree', data=strong_scaling, label='bintree', markers=True, dashes=False, marker='o')


# Customize the plot
plt.title('Strong scaling')
plt.xlabel('Size of the message (bytes)')
plt.ylabel('Time (microseconds)')
plt.legend()

# Display the plot
plt.show()
```

```python
#weak scaling MPI
weak_scaling_MPI = pd.read_csv('../MPI_data/weak_scaling_MPI_2_thin.csv')
weak_scaling_2thin = pd.read_csv('../collected_data/weak_scaling_MPI_2try.csv')

#palette
palette=sns.color_palette("bright", 3)

#drop the size column
weak_scaling_2thin = weak_scaling_2thin.drop(columns='size')
weak_scaling_MPI = weak_scaling_MPI.drop(columns='size')

# Plot the data
plt.figure(figsize=(12,  6))

#verde per bintree
sns.lineplot(x='processes', y='bintree', data=weak_scaling_2thin, label='BinTree', markers=True, dashes=False,  color=palette[2])
#blu per chain
sns.lineplot(x='processes', y='chain', data=weak_scaling_2thin, label='Chain',   markers=True, dashes=False,    color=palette[1])
#rosso per flat
sns.lineplot(x='processes', y='flat', data=weak_scaling_2thin, label='Flat',   markers=True, dashes=False,    color=palette[0])
sns.lineplot(x='processes', y='MPI', data=weak_scaling_2thin, label='MPI_Bcast',   markers=True, dashes=False,    color='red')

#verde tratteggiato per bintree
sns.lineplot(x='processes', y='bintree', data=weak_scaling_MPI, markers=True, linestyle='dashed', color=palette[2])

#blu tratteggiato per chain
sns.lineplot(x='processes', y='chain', data=weak_scaling_MPI, markers=True, linestyle='dashed', color=palette[1])

#rosso tratteggiato per flat
sns.lineplot(x='processes', y='flat', data=weak_scaling_MPI, markers=True, linestyle='dashed', color=palette[0])


# Customize the plot
plt.title('Weak scaling performance Comparison')
plt.xlabel('Number of Processes')
plt.ylabel('Time (microseconds)')
plt.legend()

# Display the plot
plt.show()

```

```python
#strong scaling MPI
strong_scaling_MPI = pd.read_csv('../MPI_data/strong_scaling_MPI_2_thin.csv')
strong_scaling_2thin = pd.read_csv('../collected_data/strong_scaling_thin_with_MPI.csv')

#palette
palette=sns.color_palette("bright", 3)

#plot the data
plt.figure(figsize=(12,  6))
sns.lineplot(x='size', y='flat', data=strong_scaling_2thin, label='flat',     markers=True, dashes=False, marker='o', color=palette[0])
sns.lineplot(x='size', y='chain', data=strong_scaling_2thin, label='chain',    markers=True, dashes=False, marker='o', color=palette[1])
sns.lineplot(x='size', y='bintree', data=strong_scaling_2thin, label='bintree', markers=True, dashes=False, marker='o', color=palette[2])
sns.lineplot(x='size', y='MPI', data=strong_scaling_2thin, label='MPI_Bcast', markers=True, dashes=False, marker='o', color='red')

#verde tratteggiato per bintree
sns.lineplot(x='size', y='tree', data=strong_scaling_MPI, markers=True, linestyle='dashed', color=palette[2])

#blu tratteggiato per chain
sns.lineplot(x='size', y='chain', data=strong_scaling_MPI, markers=True, linestyle='dashed', color=palette[1])

#rosso tratteggiato per flat
sns.lineplot(x='size', y='flat', data=strong_scaling_MPI, markers=True, linestyle='dashed', color=palette[0])

# Customize the plot
plt.title('Strong Scaling Performance Comparison')
plt.xlabel('Size of the message (bytes)')
plt.ylabel('Time (microseconds)')
plt.legend()
```
