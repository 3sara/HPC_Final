```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leggi il file CSV
s0_epyc_map_by_core = pd.read_csv('scatter_data/thin_s0_map_by_core.csv')
# Leggi il file CSV

palette = sns.color_palette('rainbow', len(s0_epyc_map_by_core['Size'].unique()))

# Imposta una figura di dimensioni appropriate
plt.figure(figsize=(10,   6))

# Crea un grafico a linee con marker per ogni valore unico nella prima colonna

for i, num in enumerate(s0_epyc_map_by_core['Size'].unique()):
    subset = s0_epyc_map_by_core[s0_epyc_map_by_core['Size'] == num]
    sns.lineplot(x='Processes', y='Latency', data=subset, color=palette[i], label=str(num), marker='o', markersize=5)

# Aggiunge le linee verticali
plt.axvline(x=12, color='gray', linestyle='--')
plt.axvline(x=24, color='gray', linestyle='--')
plt.axvline(x=36, color='gray', linestyle='--')

# Aggiunge le linee verticali
#plt.axvline(x=64, color='gray', linestyle='--')
#plt.axvline(x=128, color='gray', linestyle='--')
#plt.axvline(x=192, color='gray', linestyle='--')


plt.legend(title='Size', bbox_to_anchor=(1.05,   1), loc='upper left')

# Salva la figura
plt.savefig('scatter_plots/scatter0.png',bbox_inches='tight')
plt.show()
```
