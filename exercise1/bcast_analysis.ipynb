```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leggi il file CSV
b1_epyc_map_by_core = pd.read_csv('bcast_data/b1_epyc_map_by_core.csv')

# Genera una scala di colori blu che va dal chiaro al scuro
palette = sns.color_palette('rainbow', len(b1_epyc_map_by_core['Processes'].unique()))

# Imposta una figura di dimensioni appropriate
plt.figure(figsize=(10, 6))

# Crea un grafico a linee per ogni valore unico nella prima colonna
for i, num in enumerate(b1_epyc_map_by_core['Processes'].unique()):
    subset = b1_epyc_map_by_core[b1_epyc_map_by_core['Processes'] == num]
    sns.lineplot(x='Size', y='Latency', data=subset, color=palette[i], label=str(num))
#plt.axvline(x=8192, color='gray', linestyle='--', alpha=0.5)
#plt.axvline(x=524288, color='gray', linestyle='--', alpha=0.5)
# Mostra il grafico
#mostra la legenda accanto al grafico
plt.title('b1 map by core on epyc')
plt.legend(title='Processes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('plots/b1_epyc_map_by_core.png',bbox_inches='tight')
plt.show()
```

```python


# Genera una scala di colori blu che va dal chiaro al scuro
palette = sns.color_palette('rainbow', len(b1_epyc_map_by_core['Size'].unique()))

# Imposta una figura di dimensioni appropriate
plt.figure(figsize=(10,   6))

# Crea un grafico a linee con marker per ogni valore unico nella prima colonna

for i, num in enumerate(b1_epyc_map_by_core['Size'].unique()):
    subset = b1_epyc_map_by_core[b1_epyc_map_by_core['Size'] == num]
    sns.lineplot(x='Processes', y='Latency', data=subset, color=palette[i], label=str(num), marker='o', markersize=5)

# Aggiunge le linee verticali
plt.axvline(x=64, color='gray', linestyle='--')
plt.axvline(x=128, color='gray', linestyle='--')
plt.axvline(x=192, color='gray', linestyle='--')

# Mostra il grafico
plt.title('bcast1 per per core')
plt.legend(title='Size', bbox_to_anchor=(1.05,   1), loc='upper left')

# Salva la figura
plt.savefig('plots/b1_epyc_map_by_core_proc.png',bbox_inches='tight')
plt.show()
```

```python
fig3d=plt.figure()
ax=fig3d.add_subplot(111,projection='3d')
ax.plot_trisurf(b1_epyc_map_by_core['Processes'],b1_epyc_map_by_core['Size'],b1_epyc_map_by_core['Latency'],cmap='viridis',edgecolor='none')
ax.set_xlabel('Processes')
ax.set_ylabel('Size')
ax.set_zlabel('Latency')
ax.view_init(elev=20, azim=210)
plt.title('bcast 1')
plt.savefig('plots/b1_epyc_map_by_core_3d.png')
plt.show()
```
