```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leggi i dati dai file CSV
s0=pd.read_csv('scatter_data/scatterfix0.csv')
s1=pd.read_csv('scatter_data/scatterfix1.csv')
s2=pd.read_csv('scatter_data/scatterfix2.csv')
s3=pd.read_csv('scatter_data/scatterfix3.csv')

#rimuovo l'ultima riga da b1
s0 = s0[:-1]
s1 = s1[:-1]
s2 = s2[:-1]
s3 = s3[:-1]


# Crea un grafico a linea con 'Processes' sull'asse x e 'Latency' sull'asse y per il primo file
plt.plot(s0['Processes'], s0['Latency'], label='s0')
plt.plot(s1['Processes'], s1['Latency'], label='s1')
plt.plot(s2['Processes'], s2['Latency'], label='s2')
plt.plot(s3['Processes'], s3['Latency'], label='s3')



# Imposta i titoli degli assi
plt.xlabel('Processes')
plt.ylabel('Latency')

# Aggiungi una legenda al grafico
plt.legend()
plt.savefig("scatter_plots/scatter_fixsize")
# Mostra il grafico
plt.show()
```
